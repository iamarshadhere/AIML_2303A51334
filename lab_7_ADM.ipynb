```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = {
    'Hours Studied': [2, 4, 6, 8, 3, 5, 7, 9],
    'Marks Obtained': [40, 60, 75, 90, 50, 70, 85, 95]
}
df = pd.DataFrame(data)
print(df)
X = df['Hours Studied'].values.reshape(-1, 1)
y = df['Marks Obtained'].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='orange', label='Regression Line')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([10,20,30,40,50,60,70,80])
df=pd.DataFrame({'x':x,'y':y})
print(df)
X_train=x[:6]
Y_train=y[:6]
x_test=x[6:]
y_test=y[6:]
X_mean=np.mean(X_train)
Y_mean=np.mean(Y_train)
numerator=np.sum((X_train-X_mean)*(Y_train-Y_mean))
denominator=np.sum((X_train-X_mean)**2)
m=numerator/denominator
c=Y_mean-m*X_mean
y_pred=m*x_test+c
plt.scatter(X_train,Y_train,color='red',label='training data')
plt.scatter(x_test,y_test,color='blue',label='testing data')
plt.plot(x_test,y_pred,color='green',label='regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10, 20, 3, 40, 50, 60, 70, 80, 90, 100]).reshape(-1, 1)
Y = np.array([555, 605, 655, 705, 755, 805, 855, 905, 955, 1005])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
r_squared = r2_score(Y_test, Y_pred)

print(f"Regression Coefficients: beta_0 = {model.intercept_}, beta_1 = {model.coef_[0]}")
print("Predicted Y values:", Y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared:", r_squared)


plt.scatter(X_train, Y_train, color='blue', label='Training Data')
plt.scatter(X_test, Y_test, color='red', label='Testing Data')
plt.plot(X_test, Y_pred, color='green', linestyle='dashed', label='Regression Line')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.title("Simple Linear Regression")
plt.show()
```

```python
# Generate some data
X = np.linspace(0, 10, 100).reshape(-1, 1)  # 100 data points for X
y_true = 2 * X + 1  # True relationship (slope=2, intercept=1)

# Add noise to the data
noise_factor = 5  # Increase/decrease this value to change the noise
y_noisy = y_true + noise_factor * np.random.randn(*y_true.shape)  # Add noise

# Fit the model
model = LinearRegression()
model.fit(X, y_noisy)

# Predictions
y_pred = model.predict(X)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X, y_noisy, color='blue', label='Noisy data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.title('Effect of Noise on Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients
print("Model coefficients with noise:")
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
# Change the true slope
new_slope = 3  # Increase the slope
y_true_new_slope = new_slope * X + 1

# Fit the model to the new data
model.fit(X, y_true_new_slope)

# Predictions
y_pred_new_slope = model.predict(X)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X, y_true_new_slope, color='green', label=f'Linear data (slope={new_slope})')
plt.plot(X, y_pred_new_slope, color='red', label='Regression line')
plt.title(f'Effect of Changing Slope on Model Coefficients (Slope={new_slope})')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients
print("Model coefficients with new slope:")
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
# Use a different range for X
X_new_range = np.linspace(0, 50, 100).reshape(-1, 1)  # X range from 0 to 50

# True relationship with original slope
y_true_new_range = 2 * X_new_range + 1

# Fit the model
model.fit(X_new_range, y_true_new_range)

# Predictions
y_pred_new_range = model.predict(X_new_range)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X_new_range, y_true_new_range, color='purple', label='Data (different range for X)')
plt.plot(X_new_range, y_pred_new_range, color='red', label='Regression line')
plt.title('Effect of Different Range for X on Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients
print("Model coefficients with new X range:")
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
# Use fewer data points
X_fewer = np.linspace(0, 10, 10).reshape(-1, 1)  # Only 10 data points
y_fewer = 2 * X_fewer + 1 + noise_factor * np.random.randn(*X_fewer.shape)

# Fit the model
model.fit(X_fewer, y_fewer)

# Predictions
y_pred_fewer = model.predict(X_fewer)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X_fewer, y_fewer, color='orange', label='Fewer data points')
plt.plot(X_fewer, y_pred_fewer, color='red', label='Regression line')
plt.title('Effect of Fewer Data Points on Regression Accuracy')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients
print("Model coefficients with fewer data points:")
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
# Non-linear (quadratic) relationship
X_quadratic = np.linspace(0, 10, 100).reshape(-1, 1)
y_quadratic = 2 * X_quadratic**2 + 3 * X_quadratic + 1  # Quadratic function

# Fit the model
model.fit(X_quadratic, y_quadratic)

# Predictions
y_pred_quadratic = model.predict(X_quadratic)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X_quadratic, y_quadratic, color='brown', label='Quadratic data')
plt.plot(X_quadratic, y_pred_quadratic, color='red', label='Linear regression line')
plt.title('Effect of Non-Linear Relationship on Linear Model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients
print("Model coefficients for quadratic data:")
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Create a polynomial regression model (degree=2 for quadratic)
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())

# Fit the model
poly_model.fit(X_quadratic, y_quadratic)

# Predictions
y_pred_poly = poly_model.predict(X_quadratic)

# Plot the results
plt.figure(figsize=(10,6))
plt.scatter(X_quadratic, y_quadratic, color='brown', label='Quadratic data')
plt.plot(X_quadratic, y_pred_poly, color='blue', label='Polynomial regression line')
plt.title('Polynomial Regression on Quadratic Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
